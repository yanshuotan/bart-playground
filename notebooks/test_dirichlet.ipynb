```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import sys
import os
sys.path.append(os.path.dirname(os.getcwd()))

from bart_playground import *
```

```python
dgp = DataGenerator(n_samples=100, n_features=20, noise=0.1, random_seed=42)
X, y = dgp.sparse_linear()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.coef_
mean_squared_error(y_test, lr.predict(X_test))
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
mean_squared_error(y_test, rf.predict(X_test))
```

```python
bart = DefaultBART(dirichlet_prior=True)
bart.fit(X_train, y_train)
```

```python
bart_baseline = DefaultBART(dirichlet_prior=False)
bart_baseline.fit(X_train, y_train)
```

```python
mean_squared_error(y_test, bart_baseline.predict(X_test))
```

```python
mean_squared_error(y_test, bart.predict(X_test))
```
