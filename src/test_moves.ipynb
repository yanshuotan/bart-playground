```python
import numpy as np
import pandas as pd
num_samples = 100
num_features = 4

# 生成随机数值特征数据
X = np.random.rand(num_samples, num_features)

# 生成目标变量 y，是特征的线性组合加上随机噪声
coefficients = np.random.rand(num_features)  # 真实的系数
noise = np.random.randn(num_samples) * 0.1  # 加入一些噪声

y = X @ coefficients + noise  # 使用矩阵乘法来计算 y

# 创建 DataFrame 来保存 X 和 y
df_X = pd.DataFrame(X, columns=[f"Feature_{i+1}" for i in range(num_features)])
df_y = pd.Series(y, name="Target")

```
